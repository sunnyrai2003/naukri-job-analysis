# Salary section

```python
df['salary'].isnull()
```

```python
df['salary'].nunique()
```

```python
import numpy as np
```

```python
# Function to extract min and max salary
def extract_salary_range(salary):
    salary = salary.strip()

    if salary== 'not disclosed':
        return pd.Series([np.nan, np.nan])  # Return NaN for both min and max

    # Check for specific salary format with variable
    if '(' in salary:
        # Extract the numeric part before the parentheses
        salary = salary.split('(')[0].strip()

    # Check for salary range
    if ' - ' in salary:
        # Remove ' pa.' if present and split the range
        salary = salary.replace(' pa.', '').strip()
        min_salary, max_salary = salary.split(' - ')

        # Remove commas and check if they are numeric
        min_salary_cleaned = min_salary.replace(',', '').strip()
        max_salary_cleaned = max_salary.replace(',', '').strip()

        if min_salary_cleaned.isdigit() and max_salary_cleaned.isdigit():
            min_salary = int(min_salary_cleaned)
            max_salary = int(max_salary_cleaned)
            return pd.Series([min_salary, max_salary])
        else:
            return pd.Series([np.nan, np.nan])  # Return NaN if not valid

    # If it's a single salary value
    salary_cleaned = salary.replace(',', '').strip()
    if salary_cleaned.isdigit():
        min_salary = int(salary_cleaned)
        return pd.Series([min_salary, min_salary])  # Same value for min and max
    else:
        return pd.Series([np.nan, np.nan])  # Return NaN if not valid

# Apply the function to the salary column
df[['min_salary', 'max_salary']] = df['salary'].apply(extract_salary_range)
```
